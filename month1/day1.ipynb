## NumPy Basics

NumPy = Python library for numerical computing
              - Used for working with arrays

* Numpy is written on c. So c is faster execution than python.
* NumPy stores data in one block. So it is faster than python. Because of the Continuous memory.
* Numpy haven't loops.

Import NumPy

```python
import numpy as np 
```

Create Arrays

```python
a = np.array([1,2,3,4])
print(a)
```

```python
b = np.array([[1, 2], [3, 4]])
print(b)
```

Check Shape

```python
print(a.shape)  # Shape of array a (4,)  , represent 1D array
print(b.shape)  # Shape of array b (2, 2)  2 row 2 col
```

Operations witth array

```python
print(a + 2)
print(a * 2)
```

Create Zeros and Ones

```python
print(np.zeros((2,2)))    # 2*2 matrix filled 0
print(np.ones((3,3)))     # 3*3 matrix filled 1
```
