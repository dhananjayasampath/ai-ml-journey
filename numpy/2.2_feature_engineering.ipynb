## FEATURE ENGINEERING

* Feature Engineering - Creating NEW useful columns from existing data

```python
import pandas as pd

df = pd.DataFrame({
    "Hours": [1,2,3,4,5,6,7,8,2,6],
    "Sleep": [8,7,6,6,5,5,4,4,7,5],
    "Score": [50,55,60,65,70,75,80,85,58,78]
})

print(df)
```

Create New Feature

```python
df["Efficiency"] = df["Score"] / df["Hours"]    # Efficiency = Score per Hour

print(df)
```

```python
df["Sleep_Impact"] = df["Score"] / df["Sleep"]  # Sleep Impact Feature

print(df)
```

Conditional Feature

```python
df["Performance"] = df["Score"].apply(
    lambda x: "Good" if x > 70 else "Average"
)

print(df)
```

Data Transformation (Normalization Idea)

* Data transformation = changing data into a better format for analysis or ML
* Normalization = scaling data into a common range
* Converting data into a standard scale.So values become:
    - smaller
    - consistent
    - comparable

```python
df["Scaled_Score"] = df["Score"] / df["Score"].max()

print(df)
```

Visualization


```python
import matplotlib.pyplot as plt

plt.scatter(df["Hours"], df["Efficiency"])
plt.show()
```

```python
# Efficiency does NOT increase with hours
# It actually decreases
```
