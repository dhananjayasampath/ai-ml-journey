## REAL DATASET

```python
import pandas as pd

df = pd.DataFrame({
    "Hours": [1,2,3,4,5,6,7,8,2,6],
    "Sleep": [8,7,6,6,5,5,4,4,7,5],
    "Score": [50,55,60,65,70,75,80,85,58,78]
})

print(df)

# Create & print Dataset
```

```python
print(df.head())      # first rows
print(df.shape)       # rows, columns
print(df.info())      # data types
print(df.describe())  # statistics
```

```python
print(df.isnull().sum())

# Check Missing Data
```

```python
print(df["Score"].mean())
print(df["Hours"].mean())

#Calculates the average
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Hours"], df["Score"])
plt.xlabel("Hours")
plt.ylabel("Score")
plt.title("Hours vs Score")
plt.show()
```

* More study hours → higher score (Points go upward)
* Score increases steadily with hours
