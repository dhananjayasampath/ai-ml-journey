## REAL DATA CLEANING

```python
import pandas as pd
import numpy as np

# Create Noisy Dataset (Data that is messy or unreliable)

df = pd.DataFrame({
    "Hours": [1, 2, 3, 4, 5, 6, 7, None, 2, 6],
    "Sleep": [8, 7, 6, 6, 5, 5, 4, 4, None, 5],
    "Score": [50, 55, 60, 65, 70, 75, 80, 85, 58, 500]  # 500 is ERROR (outlier)
})

print(df)
```

```python
# Detect Missing Values
print(df.isnull().sum())
```

```python
# Handle Missing Values
df["Hours"] = df["Hours"].fillna(df["Hours"].mean())
df["Sleep"] = df["Sleep"].fillna(df["Sleep"].mean())
```

Fix Outliers

```python
# Remove extreme value
df = df[df["Score"] < 200]

# removes rows where Score is < 200
```

```python
# Recalculate Features
df["Efficiency"] = df["Score"] / df["Hours"]
df["Sleep_Impact"] = df["Score"] / df["Sleep"]

print(df)
```

```python
# Visualization
import matplotlib.pyplot as plt

plt.scatter(df["Hours"], df["Score"])
plt.show()
```
