## NumPy Indexing & Operations

### Indexing (Access Elements)

* Indexing - Access specific element in a list/array.
* Slicing - Access subset of elements.

* (*) Element wise multipliation
* np.dot() - Matrix multiplication

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
```

```python
print(arr[0])   # 10
print(arr[-1])  # 50 (Negative indexing)
```

```python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])  # 2D array
```

```python
print(matrix[0, 0])  # 1
print(matrix[1, 2])  # 6
```

### Slicing

* 1D slicing

```python
print(arr[1:4])   # [20 30 40] Between 1 and 4
print(arr[:3])    # [10 20 30] <3
print(arr[::2])   # [10 30 50] Skip 1 by 1
```

* 2D slicing

```python
print(matrix[:, 0])   # first column
print(matrix[0, :])   # first row
```

### Modify Values

```python
arr[0] = 100
print(arr)
```

### Matrix Operations

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
print(a + b)
print(a * b)
```

```python
m1 = np.array([[1, 2],
               [3, 4]])

m2 = np.array([[5, 6],
               [7, 8]])
```

```python
print(m1 + m2)
print(m1 * m2)  # element-wise
```

```python
print(np.dot(m1, m2))
```

```python
matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]
```

```python
print(matrix[1]) # Extract middle row
```

```python
print([row[2] for row in matrix]) # Extract last column
```

```python
print([[element * 3 for element in row] for row in matrix]) # Multiply entire matrix by 3
```
