## Broadcasting + Reshaping

* Broadcasting = NumPy automatically matches shapes

```python
import numpy as np

arr = np.array([1, 2, 3])
print(arr + 5)              # Adds 5 to every element without loop
```

```python
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print(a + b)           # Broadcasting with arrays
```

```python
matrix = np.array([[1,2,3],
                   [4,5,6]])

row = np.array([10, 20, 30])

print(matrix + row)         

# Different shapes
# row stretched to [10 20 30]
#                  [10 20 30]
```

Reshaping Arrays

```python
arr = np.array([1,2,3,4,5,6])
print(arr.shape)        # Original shape
```

```python
print(arr.reshape(2,3))  # array.reshape(rows, columns)
```

* No. of element must be same,

    Original elements = new rows * new columns
    *   6 = 2 * 3
    *   6 = 3 * 2

```python
print(arr.reshape(3,2))
```

Flatten

```python
matrix = np.array([[1,2,3],
                   [4,5,6]])

print(matrix.flatten())     # .flatten() converts any n-dimensional array into a 1D array
```

* reshape() → changes shape but keeps rows & columns
* flatten() → makes everything 1D
