## Pandas – Data Handling

What is a DataFrame?

* A DataFrame is a 2-dimensional labeled data structure in Python provided by pandas.
* It allows easy data manipulation, filtering, sorting, and analysis.


Why Pandas is Important in ML

* Data Cleaning & Preparation: Handles missing data, duplicates, and inconsistent formats easily.
* Data Exploration: Summarize and visualize data quickly to find patterns.
* Integration: Works well with NumPy, Matplotlib, Scikit-learn, and other ML libraries.
* Efficiency: Fast operations on large datasets using optimized functions.
* Convenience: Simple methods for filtering, grouping, merging, and reshaping data—essential for ML pipelines.

 In short, pandas makes raw data ready for machine learning models.

```python
import pandas as pd
```

Create DataFrame

```python
data = {
    "Name": ["A", "B", "C"],
    "Age": [20, 21, 22],
    "Marks": [80, 85, 90]
}                               # Python dictionary

df = pd.DataFrame(data)         # Convert dictionary to DataFrame (table)
print(df)
```

```python
print(df.head())      # Show first 5 rows
print()               # Add a blank line after each output
print(df.tail(2))     # Shows last 2 rows
print()               

print(df.shape)       # Shape (rows, columns)
print()

print(df.columns)     # Show column names
print()

print(df.info())      # Full summary of DataFrame
print()
```

Select Data

```python
print(df["Name"])        # single column
print(df[["Name","Age"]])  # multiple columns
```

```python
print(df.iloc[0])   # first row
print(df.iloc[:, 1])  # second column
```

```python
print(df.iloc[0:1, 1:2])    # Slice of rows & columns
```

```python
print(df.iloc[2, 2])    # Row i, Column j
```

Add / Modify Columns

```python
df["Grade"] = ["A", "B", "A"]   # Add new column
df["Age"] = df["Age"] + 1       # Add 1 to age column's values

print(df)
```

Filtering Data

```python
print(df[df["Marks"] > 80])
```

```python
# Marks > 80 AND Age > 21
df[(df["Marks"] > 80) & (df["Age"] > 21)]       # Multiple Conditions
```
