## Pandas Data Cleaning + Handling Missing Data

Why Clean Data?

* Real-world data often has missing values, duplicates, or errors.
* Cleaning data ensures accurate analysis and better machine learning results.
* Helps prevent errors in calculations, models, and visualizations.

* fillna() - Replaces missing values (NaN) with a value
* dropna() - Removes rows or columns with missing values

Why Use Mean/Median to Fill Missing Values?
* Mean: Average value, keeps overall data balanced. Works best for normally distributed numeric data.
* Median: Middle value, less affected by outliers. Best for skewed numeric data like salaries or prices.
* Using mean or median ensures numeric columns remain consistent for analysis or ML models.

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", None,],
    "Age": [25, None, 30, 22],
    "Salary": [50000, 60000, None, 45000]
}

df = pd.DataFrame(data)
print(df)                           # Load Dataset
```

Detect Missing Data

```python
print(df.isnull())        # Shows True where data is missing
```

```python
print(df.isnull().sum())  # Count missing values per column
```

Handle Missing Values

```python
df_drop = df.dropna()
print(df_drop)        # Drop rows with missing values
```

```python
# Fill missing values safely
df["Age"] = df["Age"].fillna(df["Age"].mean())        # Replace NaN in Age with mean
df["Salary"] = df["Salary"].fillna(df["Salary"].median())  # Replace NaN in Salary with median
df["Name"] = df["Name"].fillna("Unknown")             # Replace NaN in Name with "Unknown"

print(df)
```

```python
df.drop_duplicates(inplace=True)    # Remove duplicates
print(df)
```

```python
df["Age"] = df["Age"].astype(int)
df["Salary"] = df["Salary"].astype(float)
print(df.dtypes)            # Correct Data Types
```

🔹Practice Task

```python
data = {
    "Product": ["Pen", "Book", None, "Notebook", "Pen"],
    "Price": [10, None, 15, 20, 10],
    "Stock": [100, 200, None, 150, 100]
}
df = pd.DataFrame(data)
```

```python
df["Price"]=df["Price"].fillna(df["Price"].median())    # Fill missing numeric values with median
df["Stock"]=df["Stock"].fillna(df["Stock"].median())    
df["Product"]=df["Product"].fillna("Unknown")             # Fill missing categorical values with “Unknown”
df.drop_duplicates(inplace=True)            # Remove duplicates
print(df)
```
