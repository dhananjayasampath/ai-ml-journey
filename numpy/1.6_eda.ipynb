## EDA (Exploratory Data Analysis) + Visualization

* Pandas is a library used for data handling and analysis
* Matplotlib is used for basic data visualization (charts/graphs)
* Seaborn is used for more attractive and advanced graphs 

EDA (Exploratory Data Analysis) is the process of:
* Understanding and analyzing data before building models

What is correlation?
* Relationship between two variables
    - +1 → strong positive (both increase)
    - -1 → strong negative (one up, one down)
    -   0 → no relationship

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Basic Data Analysis

```python
df = pd.read_csv("../data/people.csv")      # Load Dataset
print(df.head())            
```

```python
print(df.describe())   # stats (mean, min, max)
print(df.value_counts())
```

```python
print(df["First Name"].value_counts())  #column-wise counts
```

Visualization

```python
plt.hist(df["Job Title"])      # Histogram (distribution)
plt.show()
```

```python
plt.scatter(df["Job Title"], df["Age"])     # Scatter Plot (relationship)
plt.show()
```

* Scatter plot = compare 2 variables visually


🔹 Why use scatter plot?
* To see relationship between two variables
* Useful to check:
    - Correlation
    - Patterns
    - Outliers

🔸 How to read it
* Points going upwards → positive relationship
* Points going downwards → negative relationship
* Random points → no clear relationship

```python
df["Age"].value_counts().plot(kind="bar")       # Bar Chart
plt.show()
```

Correlation Heatmap

* Method 1

```python
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.show()

# Correlation Heatmap
# Automatically selects all numeric columns and calculates correlation
```

* Method 2

```python
sns.heatmap(df[['Age', 'Experience']].corr(), annot=True)
plt.show()

# Correlation Heatmap
# Manually selects only specific columns
```

Method 1

 Pros:
- Fast and automatic
- Good for large datasets


Cons:
- Includes unwanted columns (like Index)
- Less control

---------------------------------------------------------------
Method 1

Pros:
- Clean and accurate
- No unnecessary columns
- Better for analysis

| Feature        | Method 1  | Method 2    |
| -------------- | ----------- | ------------- |
| Ease           | Easy        | Slight effort |
| Control        | Low         | High          |
| Accuracy       | Medium      | High          |
| Includes Index | Yes        | No           |
